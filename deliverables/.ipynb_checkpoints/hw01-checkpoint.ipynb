## Question 2a

```python
### Plot Police z-Scores
pz = pd.read_csv('./policez.csv')
pz = pz[['x']] # x column is the only one we care about
```

```python
### Plot normal distribution: N(0, 1)
mu = 0
sigma = 1
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 1000)
y = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma)**2)

plt.plot(x, y, label='Standard Normal Distribution', color = 'red')

### Plot Police z-Scores
sns.histplot(pz['x'], label = 'Police Distribution', stat = 'density');
sns.kdeplot(x=pz['x'], color='black'); #Plot the KDE

pz_mean = np.mean(pz['x'])

plt.axvline(x=pz_mean, color='green', linestyle='--', label=f'Police Distribution Mean: z = {round(pz_mean, 2)}')

plt.title('Police Distribution v. Standard Normal Distribution')
plt.xlabel('z-scores')
plt.ylabel('p(z)')
plt.grid(False)

plt.legend()
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show();
```

**<ins>Answer</ins>:** The theorhetical null does not fit the data exactly. This is because the pdf of the data has a larger standard deviation and is centered slightly to the right (0.8) of the theorhetical mean (0).

## Question 2b

```python
### Get the p-values
p_values = norm.cdf([-z for z in pz['x']]) #norm.cdf(-z) gets the p values for each of the z-scores
pz['p'] = p_values

### Apply the BH procedure with α = 0.2.
alpha = 0.2
pvs = pz[['p']]
pvs = pvs.sort_values('p')

m = len(pvs)

pvs['k'] = np.arange(1, m+1)
pvs['b-h guide'] = pvs['k'] * alpha/ m

try:
    bh_threshold = max(pvs[pvs['p'] < pvs['b-h guide']]['p'])
except: # In case none of the p-values are less than the b-h guide value
    bh_threshold = 0


sns.scatterplot(data = pvs, x='k', y='p', color = 'blue')
plt.axhline(alpha, label=f'Normal Threshold: {alpha}', color='black')
plt.axhline(bh_threshold, label=f'B-H Threshold: {round(bh_threshold, 2)}', color='red')
plt.title('Sorted p-values with Different Thresholds')
plt.legend()
plt.show();
```

```python
normal_num_discoveries = np.sum(pvs['p'] < alpha)
bh_num_discoveries = np.sum(pvs['p'] < bh_threshold)
print(f'Number of Discoveries (Normal Threshold): {normal_num_discoveries}\n\
Number of Discoveries (B-H Threshold): {bh_num_discoveries}')
```

**<ins>Answer</ins>:** Using the normal threshold, I made 833 discoveries. However, using the threshold obtained from applying the BH procedure with α = 0.2, I only made 200 discoveries.

## Question 2c

```python
### Plot normal distribution: N(0.1, 1.4^2)
mu = 0.1
sigma = 1.4**2
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 1000)
y = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma)**2)

plt.plot(x, y, label='Empirical Distribution', color = 'red')

### Plot Police z-Scores
sns.histplot(pz['x'], label = 'Police Distribution', stat = 'density');
sns.kdeplot(x=pz['x'], color='black'); #Plot the KDE

pz_mean = np.mean(pz['x'])

plt.axvline(x=pz_mean, color='green', linestyle='--', label=f'Police Distribution Mean: z = {round(pz_mean, 2)}')

plt.title('Police Distribution v. Empirical Distribution')
plt.xlabel('z-scores')
plt.ylabel('p(z)')
plt.grid(False)

plt.legend()
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show();
```

**<ins>Answer</ins>:** The empirical null does not fit the data exactly, but it is very close. Despite having a center 0.02 off of the empirical distribution (0.08 instead of 0.1), and smaller standard deviation, all the data is falls within the empirical distribution.

```python
### Get the p-values
mean = 0.10
std = 1.4**2

p_values = norm.cdf([-z for z in pz['x']], loc = mean, scale = std) #norm.cdf(-z) gets the p values for each of the z-scores
pz['p'] = p_values

### Apply the BH procedure with α = 0.2.
alpha = 0.2
m = len(pvs)
pvs = pz[['p']]
pvs = pvs.sort_values('p')
pvs['k'] = np.arange(1, m+1)
pvs['b-h guide'] = pvs['k'] * alpha/ m

try:
    bh_threshold = max(pvs[pvs['p'] < pvs['b-h guide']]['p'])
except: # In case none of the p-values are less than the b-h guide value
    bh_threshold = 0


sns.scatterplot(data = pvs, x='k', y='p', color = 'blue')
plt.axhline(alpha, label=f'Normal Threshold: {alpha}', color='black')
plt.axhline(bh_threshold, label=f'B-H Threshold: {round(bh_threshold, 2)}', color='red')
plt.title('Sorted p-values with Different Thresholds')
plt.legend()
plt.show();
```

By using the empirical distribution, which much better fits the data, it makes it much more difficult to reject the null. This is why I made much fewer discoveries than when using the standard normal distribution.

# Question 3

```python
adults = pd.read_csv('./adult.csv')
adults = adults.iloc[:, 2:] # these are the columns we care about
```

## Question 3a

```python
def difference_of_means(tbl, binary_col, numeric_col):
    relevant = tbl[[binary_col, numeric_col]] # Only selects the columns needed
    means_tbl = relevant.groupby(binary_col).agg(np.mean) 
    means = means_tbl.iloc[:, 0]
    try:
        diff = means[1] - means[0]
        return diff
    except: # If for some reason, the binary_col for the tbl is all 1's or all 0's. If only one category, can't be a difference
        return 0

def single_difference_in_means(binary_col, numeric_col):
    shuffled_labels = adults.loc[:, binary_col].sample(n = len(adults), replace=False).values
    shuffled_tbl = pd.DataFrame()
    shuffled_tbl[numeric_col] = adults[numeric_col]
    shuffled_tbl[binary_col] = shuffled_labels
    return difference_of_means(shuffled_tbl, binary_col, numeric_col) 

def avg_difference_in_means(binary_col, numeric_col):
    observed_difference = difference_of_means(adults, binary_col, numeric_col)
    
    differences = []

    repetitions = 25000
    for i in np.arange(repetitions):
        new_difference = single_difference_in_means(binary_col, numeric_col)
        differences += [new_difference]
    
    if observed_difference > 0:
        empirical_p = np.mean([differences >= observed_difference])
    else:
        empirical_p = np.mean([differences <= observed_difference])
    return empirical_p
```

## Question 3b

```python
np.random.seed(29)
binary_cols = adults[['Post HS?', 'Ever Married?', 'Gender', '50K?']].columns
numeric_cols = adults[['Age', 'Hours per Week']].columns

p_vals = pd.DataFrame(columns = ['binary_col', 'numeric_col', 'p'])

for binary_col in binary_cols:
    for numeric_col in numeric_cols:
        p = avg_difference_in_means(binary_col, numeric_col)
        p_vals.loc[len(p_vals)] = {'binary_col' : binary_col,
                                   'numeric_col' : numeric_col,
                                   'p' : p}

p_vals
```

## Question 3c

**<ins>Answer</ins>:** Using a naive p-value threshold of 0.05 to make a decision for each hypothesis test we reject the null for the tests below-

```python
alpha = 0.05
naive_rejections = p_vals[p_vals['p'] <= 0.05].reset_index(drop=True)
naive_rejections
```

## Question 3d

```python
def bonferroni_correction(alpha, df):
    m = len(df)
    result_df = df[df['p'] <= alpha / m].sort_values('p').reset_index(drop = True)
    threshold = alpha / m
    print(f'Threshold: {round(threshold, 5)}')
    return result_df, threshold
```

**<ins>Answer</ins>:** To guarantee a Family-wise Error Rate (FWER) of 0.05 when making a decision for each hypothesis test we reject the null for the tests below-

```python
alpha = 0.05
bc_rejections = bonferroni_correction(alpha, p_vals)
bc_rejections[0]
```

## Question 3e

```python
def bh_algorithm(alpha, df):
    m = len(df)
    sorted_df = df.sort_values('p')
    sorted_df['k'] = np.arange(1, m+1)
    sorted_df['b-h guide'] = sorted_df['k'] * alpha/ m

    try:
        bh_threshold = max(sorted_df[sorted_df['p'] < sorted_df['b-h guide']]['p'])
    except: # In case none of the p-values are less than the b-h guide value
        bh_threshold = 0
    
    result_df = sorted_df[sorted_df['p'] <= bh_threshold].reset_index(drop = True)
    print(f'Threshold: {round(bh_threshold, 5)}')
    return result_df, bh_threshold
```

**<ins>Answer</ins>:** To guarantee  a False Discovery Rate (FDR) of 0.05 when making a decision for each hypothesis test we reject the null for the tests below-

```python
alpha = 0.05
bh_rejections = bh_algorithm(alpha, p_vals)
bh_rejections[0]
```

## Question 3f

```python
alpha = 0.05
m = len(p_vals)
p_vals = p_vals.sort_values('p').reset_index(drop=True)
p_vals['k'] = np.arange(1, m+1)
p_vals['b-h guide'] = p_vals['k'] * alpha/ m

sns.scatterplot(data = p_vals, x='k', y='p', color = 'blue')
plt.axhline(alpha, label=f'Normal Threshold: {alpha}', color='black')
plt.axhline(bc_rejections[1], label=f'Bonferroni Threshold: {round(bc_rejections[1], 5)}', color='green')
plt.plot(p_vals['k'], p_vals['b-h guide'], label=f'B-H Guide', color='cyan')
plt.axhline(bh_rejections[1], label=f'B-H Threshold: {round(bh_rejections[1], 5)}', color='red')
plt.title('Sorted p-values with Different Thresholds')
plt.legend()
plt.show();
```

**<ins>Answer</ins>:** The results in (d) and (e) slightly differ from one another because while the results in (d) only reject the null hypothesis in three tests - (Ever Married?, Age), (50K?, Age), (Gender, Hours per Week) - (e) additionally rejects the null for another test, (Gender, Age). This difference stems from the fact that the threshold from (d), using the Bonferroni correction, is much lower (0.00625) than the threshold from (e), using the Benjamini-Hochberg algorithm (0.02136). The Bonferroni correction leads to a much more conservative test that gives us more control over our False Discovery Rate. However, at the same time, it additionally makes is much more difficult to reject the null, which is why there are less discoveries in (d).

**<ins>Answer</ins>:** For the 'Education' column, an example of another way that we could binarize the variable is to one-hot-encode it. In the adult.csv file, the 'Education' column was binarized through the 'Post HS?' column. However, once examing the 'Education' column of the adult_original.csv more in depth, it becomes apparent that there are many different levels of education (e.g. Bachelors, Preschool, etc.). Therefore, by one-hot-encoding the column, we could be binarizing the column for every unique different level of education (i.e. we would have a binary categorical column for whether or not a person is attending Preschool, completing their Doctorate, etc.). This would allow us to use our empirical p-value to determine whether or not there is a significant difference in a given numeric column between different levels of education, instead of merely between college-educated and non-college-educated adults. Such a change in the binarization strategy would be quite powerful as the empirical p-values would give us much more helpful insights than we'd receive from the rudimentary binarization in the adult.csv file.
